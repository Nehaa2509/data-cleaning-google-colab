```
import pandas as pd

house_df = pd.read_excel("/content/.config/House_Prices_Dataset.xlsx")
house_df
```

```
house_df.isnull().sum()
```

```
# Fill numerical missing values with median
house_df['LotFrontage'].fillna(house_df['LotFrontage'].median(), inplace=True)

# Verify
house_df.isnull().sum()
```

```
medical_df = pd.read_excel("/content/.config/Medical_Appointment_No_Shows.xlsx")
medical_df
```

```
medical_df = medical_df[medical_df['Age'] >= 0]
```

```
medical_df['ScheduledDay'] = pd.to_datetime(medical_df['ScheduledDay'])
medical_df['AppointmentDay'] = pd.to_datetime(medical_df['AppointmentDay'])
```

```
medical_df['No-show'] = medical_df['No-show'].map({'No': 0, 'Yes': 1})
```

```
house_df.to_excel("/content/.config/House_Prices_Dataset.xlsx", index=False)
medical_df.to_excel("/content/.config/Medical_Appointment_No_Shows.xlsx", index=False)
```
